1.given a string "exercicio tpc1", reverse it.

```python
s ="Exercicio tcp1"
def reverse_string(s):
    return s[::-1]
print(reverse_string(s))
```

2.given a string "s", returns how many "a" and "A" characters are present in it.

```python
s = "Ana Maria Lima"

def count_a(s):
    count_lower = s.count('a')
    count_upper = s.count('A')
    return count_lower, count_upper

lower, upper = count_a(s)

print("a:", lower)
print("A:", upper)
```

3.given a string "s", returns the number of vowels there are present in it.

```python
s = "AAAeieoaiiadiaihsafafakhhasaaiiiiio"
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

print("Number of vowels:", count_vowels(s))
```

```python
s = "Exercicio tcp1"
def is_capicua(s):
    return s == s[::-1]
print(is_capicua(s)) 
```

7.Verifica se duas strings estão balanceadas (Duas strings, s1 e s2, estão balanceadas se todos os caracteres de s1 estão presentes em s2.)

```python
s1 = "abcde"
s2 = "edcba"
def balanceadas(s1, s2):
    for char in s1:
        if char not in s2:
            return False
    return True

print(balanceadas(s1, s2))
```

9.Verifica se s1 é anagrama de s2.

->"listen" e "silent": Deve imprimir True

```python
s1 = "listen"
s2 = "silent"

def is_anagram(s1, s2):
    return sorted(s1) == sorted(s2)

print(is_anagram(s1, s2))   
```
